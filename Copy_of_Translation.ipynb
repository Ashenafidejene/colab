```python
import numpy as np
import imutils
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('trex.png')
```

```python
cv2_imshow(image)
```

```python
m = np.float32([[1,0,25],[0,1,5]])
shifted = cv2.warpAffine(image,m,(image.shape[1],image.shape[0]))
print(shifted)
cv2_imshow(shifted)
```

```python
(h,w) = image.shape[:2]
center = (w//2,h//2)
m = cv2.getRotationMatrix2D(center,-45,1.0)
rotated = cv2.warpAffine(image,m,(w,h))
cv2_imshow(rotated)
```

```python
fliped = cv2.flip(image,)
cv2_imshow(fliped)
```
